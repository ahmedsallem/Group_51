## Pandas Series:

```python
import numpy as np
import pandas as pd
arr = np.array([1,2,3])
row_names = ['a','b','c']
myseries = pd.Series(arr , index = row_names)
print(myseries)
print(myseries[0])
print(myseries['a'])
```

## Pandas DataFrame:

```python
myarr = np.array([[1,2,3],[3,4,5]])
row_names = ['a','b']
col_names = ['one','two','three']
my_df = pd.DataFrame(myarr , index = row_names ,columns= col_names)
print(my_df)
print("method_1")
print(my_df['one'])
print("method_2")
print(my_df.one)
```

```python
# Handle missing values by dropping 
# rows with missing data or using other way
df.dropna(inplace=True)
# Check again missing values
df.isnull().sum()
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
# Drop duplicates
df.drop_duplicates(inplace=True)
# check again
df.duplicated().sum()
```

```python
# Identify outliers using z-score
from scipy.stats import zscore
df['z_score'] = zscore(df['price'])
df = df.loc[df['z_score'].abs() < 3]
# remove z-score column
df.drop('z_score', axis=1, inplace=True)
```

```python
df.tail()
```

```python
# drop the columns irrelevant to our analysis
df.drop(columns = ['id','host_id','host_name','latitude','longitude','calculated_host_listings_count'],
           axis = 1,inplace = True)
df.columns
```

```python
# Export the cleaned data
df.to_csv('cleaned_data.csv', index=False)
```
