```python
import numpy as np
import pandas as pd
```

```python
f = pd.read_csv("PlayTennis.csv")
```

```python
df = pd.DataFrame(f)
df
```

```python
total = df["Play Tennis"].count()
num_yes = len(df[df["Play Tennis"]=="Yes"])
num_no = len(df[df["Play Tennis"]=="No"])
probabiltyYes = num_yes / total
probabiltyNo = num_no / total
```

```python

def likelihoodTables(df):
    likelihood_tables = {}
    for col in df.columns:
        if col == "Play Tennis":
            continue

        likelihood_table = {}

        for value in df[col].unique():
            counts = df[df[col] == value]["Play Tennis"].value_counts()
            probabilities = {}

            for play, count in counts.items():
                if play == "Yes":
                    probabilities[play] = count / num_yes
                else:
                    probabilities[play] = count / num_no

            likelihood_table[value] = probabilities

        likelihood_tables[col] = likelihood_table

    return likelihood_tables


def predict(test_sample, likelihood_tables,probYes,probNo):
    probabilities = {}
    yes_prob = probYes
    no_prob = probNo
    for col, value in test_sample.items():
        if col in likelihood_tables:
            likelihood_table = likelihood_tables[col]
            if value in likelihood_table:
                probabilities[col] = likelihood_table[value]

    if not probabilities:
        return None
    
    for prob in probabilities.values():
        yes_prob *= prob.get("Yes", 0)
        no_prob *= prob.get("No", 0)
        
    print (yes_prob,no_prob)
    if yes_prob > no_prob:
        return "Yes"
    else:
        return "No"




test_sample = {
    "Outlook": "Sunny",
    "Temperature": "Cool",
    "Humidity": "High",
    "Wind": "Strong"
}
test_sample = {}

# for col in df.columns:
#     if col != "Play Tennis":
#         value = input(f"Enter {col}: ")
#         test_sample[col] = value
likelihood_tables = likelihoodTables(df)        
prediction = predict(test_sample, likelihood_tables,probabiltyYes,probabiltyNo)

print("Prediction:", prediction)
```
